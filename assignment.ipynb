```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns



df.hist(figsize=(10, 8), bins=15)
plt.suptitle("Histogram ", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='SL', y='SW', hue='Classification', data=df, s=80)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Classification', y='SL', data=df)
plt.title("Boxplot of Sepal Length by Species")
plt.show()
```

```python
sns.pairplot(df, hue='Classification')
plt.suptitle("Pairplot of Iris Dataset", fontsize=16)
plt.show()
```
