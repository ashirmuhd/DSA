2.Sales dataset

```python
import pandas as pd
import math

df = pd.read_csv("Sales_add.csv")

print(df.head(), "\n")

before = df['Sales_before_digital_add(in $)']
after = df['Sales_After_digital_add(in $)']

# Calculate differences
diff = after - before

# Mean of differences
mean_diff = diff.mean()

# Standard deviation of differences
std_diff = diff.std(ddof=1)

# Number of observations
n = len(diff)

# Calculate t-statistic
t_stat = mean_diff / (std_diff / math.sqrt(n))

print("Paired t-test (Manual Calculation):")
print("Mean Difference:", mean_diff)
print("Standard Deviation of Difference:",std_diff)
print("t-statistic:",t_stat)

# Decision at significance level alpha = 0.05
# Critical t-value for df = n-1 ≈ 2.093 (approx for 10 samples)
critical_t = 2.093

if abs(t_stat) > critical_t:
    print("Conclusion: Sales increased significantly after digital marketing.\n")
else:
    print("Conclusion: No significant change in sales.\n")
```

```python

#b

cont_table = pd.crosstab(df['Region'], df['Manager'])
print("Contingency Table:\n", cont_table, "\n")

# Calculate expected values
row_totals = cont_table.sum(axis=1)
col_totals = cont_table.sum(axis=0)
grand_total = cont_table.values.sum()

chi_square = 0

for i in cont_table.index:
    for j in cont_table.columns:
        expected = (row_totals[i] * col_totals[j]) / grand_total
        observed = cont_table.loc[i, j]
        chi_square += ((observed - expected) ** 2) / expected

print(f"Chi-Square Value (Manual Calculation): {chi_square:.4f}")

# Critical chi-square value for df = (rows-1)*(cols-1)
# For 3 regions, 3 managers → df = 4
# Critical value ≈ 9.488 at α = 0.05
critical_chi = 9.488

if chi_square > critical_chi:
    print("Conclusion: Region and Manager are dependent (related).")
else:
    print("Conclusion: Region and Manager are independent (not related).")
```
