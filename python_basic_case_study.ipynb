1.Bank Account Manager

```python
class BankAccount:
    def __init__(self):
        self.balance = 0

    def deposit(self):
        amount = int(input("Enter amount to deposit: "))
        self.balance = self.balance + amount
        print("Amount Deposited!")

    def withdraw(self):
        amount = int(input("Enter amount to withdraw: "))
        if amount <= self.balance:
            self.balance = self.balance - amount
            print("Amount Withdrawn!")
        else:
            print("Not enough balance!")

    def show_balance(self):
        print("Current Balance:", self.balance)

# Main Program
account = BankAccount()

while True:
    print("\n1. Deposit\n2. Withdraw\n3. Balance Check\n4. Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        account.deposit()
    elif choice == 2:
        account.withdraw()
    elif choice == 3:
        account.show_balance()
    elif choice == 4:
        print("Thank you!")
        break
    else:
        print("Invalid Option")
```

2.Weather Data Analyzer

```python
temperature = [30, 32, 29, 35, 31]
humidity = [60, 65, 70, 75, 68]

def avg_temp(temp):
    return sum(temp) / len(temp)

def avg_hum(hum):
    return sum(hum) / len(hum)

def condition(t):
    if t > 33:
        return "Hot"
    elif t < 29:
        return "Cold"
    else:
        return "Normal"

print("Average Temperature:", avg_temp(temperature))
print("Average Humidity:", avg_hum(humidity))

today = temperature[-1]
print("Today's Weather Condition:", condition(today))
```
