##bar

```python
products = ["Books","Charts","Pens","Eraser","Paper"]
count = [200,25,150,35,135]
plt.bar(products,count,color ='lightgreen')
plt.show()
```

```python
products = ["Books","Charts","Pens","Eraser","Paper"]
count = [200,25,150,35,135]
plt.barh(products,count,color ='c')
plt.xlabel('count')
plt.ylabel('products')
plt.title('Sales')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]
colors = ['red', 'green', 'blue', 'orange']

plt.bar(categories, values, color=colors, edgecolor='black', width=0.6)

plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Colored Bar Plot with Edge Lines')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

##PIE


```python
products = ["Books","Charts","Pens","Eraser","Paper"]
count = [200,25,150,35,135]
explode = [0, 0.1, 0, 0, 0]  # Explode the second slice (Charts)
plt.pie(count, labels=products, explode=explode)
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
sizes = [20, 50, 100, 200, 80, 30, 300, 60, 90, 150]  # marker sizes
colors = ['red', 'green', 'blue', 'purple', 'orange', 'cyan', 'black', 'pink', 'gray', 'lime']

plt.scatter(x, y, s=sizes, c=colors, alpha=0.9, edgecolor='black')

plt.title('Custom Scatter Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(True)
plt.show()
```

##Seaborn

```python
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.scatterplot(data = iris,x ='sepal_length',y = 'sepal_width',hue = 'species')
plt.title('Sepal Length VS Sepal Width')
```

```python
sns.scatterplot(data = iris,x = 'petal_length' , y = 'petal_width',hue = 'species')
plt.title('Petal Length VS Petal Width')
```

```python
sns.pairplot(data = iris,hue = 'species')
```

```python
sns.boxplot(data = iris,x ='species',y = 'petal_length')
```

```python
#violinplot,countplot,stripplot,swarmplot,kdeplot
```

```python
iris_num = iris.select_dtypes(include=['int64','float64'])
iris_num.head()
```

```python
corr_matrix = iris_num.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix ,annot=True,cmap='plasma')
```

```python
#violinplot
sns.violinplot(x='species', y='sepal_length', data=iris,hue = 'species')
plt.title("Violin Plot of Sepal Length by Species")
plt.show()
```

```python
#countplot
sns.countplot(x="species", data=iris)
plt.title("Count of Each Iris Species")
plt.show()
```

```python
#stripplot
sns.stripplot(x="species", y="petal_length", data=iris,hue='species')
plt.title("Strip Plot of Petal Length by Species")
plt.show()
```

```python
#swarmplot
sns.swarmplot(x="species", y="petal_width", data=iris)
plt.title("Swarm Plot of Petal Width by Species")
plt.show()

```

```python
#kdeplot
sns.kdeplot(data=iris, x="sepal_length", y="sepal_width")#, fill=True)
plt.title("KDE Plot of Sepal Length by sepal width")
plt.show()
```
